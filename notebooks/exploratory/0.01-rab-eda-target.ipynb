# EDA on Target Variable: exam_score

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Adjust path for notebook location
train_df = pd.read_csv('../../data/raw/train.csv')

target_col = 'exam_score'
```

## 1. Descriptive Statistics

```python
print(train_df[target_col].describe())
```

## 2. Missing Values

```python
print(f"Missing values in {target_col}: {train_df[target_col].isnull().sum()}")
```

## 3. Distribution Visualization

```python
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(train_df[target_col], kde=True)
plt.title(f'Distribution of {target_col}')
plt.xlabel(target_col)
plt.ylabel('Frequency')

# Box Plot
plt.subplot(1, 2, 2)
sns.boxplot(x=train_df[target_col])
plt.title(f'Box Plot of {target_col}')
plt.xlabel(target_col)

plt.tight_layout()
plt.show()
```

## 4. Skewness and Kurtosis

```python
print(f"Skewness: {train_df[target_col].skew()}")
print(f"Kurtosis: {train_df[target_col].kurt()}")
```
